### Q1 - Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos

```python

# Set up the Selenium driver
driver = webdriver.Chrome()  # Use Chrome driver, you can choose other drivers as well
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")  # Navigate to the Wikipedia URL

# Create empty lists to store the scraped data
rank_list = []
name_list = []
artist_list = []
upload_date_list = []
views_list = []

# Find the table element containing the most viewed videos
try:
    table = driver.find_element(By.CLASS_NAME, "wikitable")
    rows = table.find_elements(By.TAG_NAME, "tr")
    for row in rows[1:]:
        # Extract data from each row
        cells = row.find_elements(By.TAG_NAME, "td")
        if len(cells) >= 5:
            rank_list.append(cells[0].text.strip())
            name_list.append(cells[1].text.strip())
            artist_list.append(cells[2].text.strip())
            upload_date_list.append(cells[4].text.strip())
            views_list.append(cells[3].text.strip())

except NoSuchElementException as e:
    print("Element not found: ", e)

finally:
    # Close the Selenium driver
    driver.quit()

# Print the scraped data
print("Rank: ", rank_list)
print("Name: ", name_list)
print("Artist: ", artist_list)
print("Upload Date: ", upload_date_list)
print("Views: ", views_list)
```

### Q2 - Scrape the details of team Indiaâ€™s international fixtures from bcci.tv. 

```python


# Set up the Selenium driver
driver = webdriver.Chrome()  # Use Chrome driver, you can choose other drivers as well
driver.get("https://www.bcci.tv/")  # Navigate to the BCCI.tv home page

# Click on the "International" link to go to the international fixtures page
try:
    international_link = driver.find_element(By.XPATH, "//a[contains(text(),'INTERNATIONAL')]")
    international_link.click()
    time.sleep(2)  # Add a delay to allow the page to load

except NoSuchElementException as e:
    print("Element not found: ", e)

# Create empty lists to store the scraped data
match_title_list = []
series_list = []
place_list = []
date_list = []
time_list = []

# Find the fixture elements
try:
    fixtures = driver.find_elements(By.ID, "match-card")
    for fixture in fixtures:
        match_title = fixture.find_element(By.XPATH, "//h5[@class='match-tournament-name ng-binding']").text
        #series = fixture.find_element(By.CLASS_NAME, "fixture__description__event").text
        place = fixture.find_element(By.XPATH, "//span[@class='ng-binding']").text
        date = fixture.find_element(By.XPATH, "//div[@class='match-dates ng-binding']").text
        time = fixture.find_element(By.XPATH, "//div[@class='match-time no-margin ng-binding']").text

        match_title_list.append(match_title)
        #series_list.append(series)
        place_list.append(place)
        date_list.append(date)
        time_list.append(time)

except NoSuchElementException as e:
    print("Element not found: ", e)

finally:
    # Close the Selenium driver
    driver.quit()

# Print the scraped data
print("Match Title: ", match_title_list)
print("Series: ", series_list)
print("Place: ", place_list)
print("Date: ", date_list)
print("Time: ", time_list)
```

### Scrape the details of top 100 songs on billiboard.com. Url = https:/www.billboard.com/

```python


# Set up the Chrome driver
driver = webdriver.Chrome()

# Navigate to the Billboard Hot 100 page
driver.get("https://www.billboard.com/charts/hot-100")

try:
    # Wait for the page to load
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//li[contains(@class,'chart-list__element')][1]")))

    # Scrape the details of each song
    songs = driver.find_elements(By.XPATH, "//li[contains(@class,'chart-list__element')]")

    for song in songs:
        # Get the song name
        song_name = song.find_element(By.XPATH, ".//span[contains(@class,'chart-element__information__song')]").text

        # Get the artist name
        artist_name = song.find_element(By.XPATH, ".//span[contains(@class,'chart-element__information__artist')]").text

        # Get the last week rank
        last_week_rank = song.find_element(By.XPATH, ".//div[contains(@class,'chart-element__meta text--center color--secondary text--last')]").text

        # Get the peak rank
        peak_rank = song.find_element(By.XPATH, ".//div[contains(@class,'chart-element__meta text--center color--secondary text--peak')]").text

        # Get the weeks on board
        weeks_on_board = song.find_element(By.XPATH, ".//div[contains(@class,'chart-element__meta text--center color--secondary text--week')]").text

        # Print the song details
        print(f"Song Name: {song_name}")
        print(f"Artist Name: {artist_name}")
        print(f"Last Week Rank: {last_week_rank}")
        print(f"Peak Rank: {peak_rank}")
        print(f"Weeks on Board: {weeks_on_board}")
        print("")

except Exception as e:
    print(e)

finally:
    # Quit the driver
    driver.quit()
```

```python
from selenium import webdriver
import time

# create webdriver object
driver = webdriver.Chrome()

# navigate to naukri.com homepage
driver.get("https://www.naukri.com/")

# locate and click on the recruiters option
recruiters = driver.find_element(By.XPATH, "//a[@title='Search Recruiters']")
recruiters.click()

# switch to new window
driver.switch_to.window(driver.window_handles[1])

# locate and fill the search pane with "Data science"
search_box = driver.find_element(By.XPATH, "//input[@class='sugInp']")
search_box.send_keys("Data science")

# click on search button
search_button = driver.find_element(By.XPATH, "//button[@id='qsbFormBtn']")
search_button.click()

# wait for the results to load
time.sleep(5)

# scrape the details of top 10 recruiters
recruiters = driver.find_elements(By.XPATH, "//div[@class='recInfo']")[:10]

for recruiter in recruiters:
    # scrape details of each recruiter
    name = recruiter.find_element(By.XPATH, ".//div[@class='recName']/span/a").text
    designation = recruiter.find_element(By.XPATH, ".//div[@class='recDesign']/span").text
    try:
        company = recruiter.find_element(By.XPATH, ".//div[@class='recInfo']/div/p/a[1]").text
    except:
        company = None
    skills = recruiter.find_element(By.XPATH,".//div[@class='recInfo']/div[2]").text
    location = recruiter.find_element(By.XPATH, ".//div[@class='recInfo']/div[3]/span").text
    
    # print the details
    print("Name:", name)
    print("Designation:", designation)
    print("Company:", company)
    print("Skills they hire for:", skills)
    print("Location:", location)
    print("-" * 50)

# close the webdriver
driver.quit()
```
